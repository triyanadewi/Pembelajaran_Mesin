### **Tugas 1**
Terdapat dataset mushroom. Berdasarkan dataset yang tersebut, bandingkan peforma antara algoritma Decision Tree dan RandomForest. Gunakan tunning hyperparameter untuk mendapatkan parameter dan akurasi yang terbaik.

**Tugas 1: Perbandingan Performa Decision Tree dan RandomForest dengan Dataset Mushroom**

**1. Import Library**

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
```

**2. Persiapan Data**
<br>Load dataset dan lakukan preprocessing, seperti mengencode label kategori dan memisahkan fitur dan target.

```python
# Load dataset
df = pd.read_csv('data/mushrooms.csv')

# Menampilkan 5 data teratas
df.head()
```

```python
# Mengecek jumlah nilai null
df.isnull().sum()
```

```python
# Encoding label
label_encoder = LabelEncoder()
df_encoded = df.apply(label_encoder.fit_transform)

# Memisahkan fitur dan target
X = df_encoded.drop('class', axis=1)
y = df_encoded['class']

# Memisahkan data menjadi data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

**3. Training Decision Tree dengan Hyperparameter Tuning**
<br>Melakukan hyperparameter tuning untuk mendapatkan parameter terbaik.

```python
# Parameter grid untuk Decision Tree
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Membuat objek Decision Tree
dt = DecisionTreeClassifier(random_state=42)

# Grid Search untuk hyperparameter tuning
grid_search_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X_train, y_train)

# Hasil hyperparameter terbaik
print("Best parameters for Decision Tree:", grid_search_dt.best_params_)

# Memprediksi dengan model terbaik
y_pred_dt = grid_search_dt.predict(X_test)

# Menghitung akurasi
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Akurasi Decision Tree: {:.4f}".format(acc_dt))
print(classification_report(y_test, y_pred_dt))
```

**4. Training Random Forest dengan Hyperparameter Tuning**
<br>Melakukan hyperparameter tuning untuk model Random Forest.

```python
# Parameter grid untuk Random Forest
param_grid_rf = {
    'n_estimators': [10, 50, 100],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Membuat objek Random Forest
rf = RandomForestClassifier(random_state=42)

# Grid Search untuk hyperparameter tuning
grid_search_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)

# Hasil hyperparameter terbaik
print("Best parameters for Random Forest:", grid_search_rf.best_params_)

# Memprediksi dengan model terbaik
y_pred_rf = grid_search_rf.predict(X_test)

# Menghitung akurasi
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Akurasi Random Forest: {:.4f}".format(acc_rf))
print(classification_report(y_test, y_pred_rf))
```

**Tugas 2: Perbandingan Performa Decision Tree dan AdaBoost pada Dataset Mushroom**

**1. Import library tambahan yaitu Ada Boost Classifier**

```python
from sklearn.ensemble import AdaBoostClassifier
```

**2. Training Decision Tree dengan Hyperparameter Tuning**
<br>
Seperti pada tugas 1, menggunakan Decision Tree dengan hyperparameter tuning untuk mendapatkan parameter terbaik.

```python
# Parameter grid untuk Decision Tree
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Membuat objek Decision Tree
dt = DecisionTreeClassifier(random_state=42)

# Grid Search untuk hyperparameter tuning
grid_search_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X_train, y_train)

# Hasil hyperparameter terbaik
print("Best parameters for Decision Tree:", grid_search_dt.best_params_)

# Memprediksi dengan model terbaik
y_pred_dt = grid_search_dt.predict(X_test)

# Menghitung akurasi
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Akurasi Decision Tree: {:.4f}".format(acc_dt))
print(classification_report(y_test, y_pred_dt))
```

<br>Pada bagian ini, akan dibangun model AdaBoost. AdaBoost memerlukan estimator, dan dalam hal ini menggunakan Decision Tree sebagai base estimator. Kemudian, dilakukannya hyperparameter tuning untuk AdaBoost.

```python
# Parameter grid untuk AdaBoost
param_grid_ada = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}

# Membuat objek AdaBoost dengan Decision Tree sebagai base estimator
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(), algorithm='SAMME', random_state=42)

# Grid Search untuk hyperparameter tuning
grid_search_ada = GridSearchCV(ada, param_grid_ada, cv=5, scoring='accuracy')
grid_search_ada.fit(X_train, y_train)

# Hasil hyperparameter terbaik
print("Best parameters for AdaBoost:", grid_search_ada.best_params_)

# Memprediksi dengan model terbaik
y_pred_ada = grid_search_ada.predict(X_test)

# Menghitung akurasi
acc_ada = accuracy_score(y_test, y_pred_ada)
print("Akurasi AdaBoost: {:.4f}".format(acc_ada))
print(classification_report(y_test, y_pred_ada))
```

### **Tugas 3**

Dengan menggunakan dataset diabetes, buatlah ensemble voting dengan algoritma

1. Logistic Regression
2. SVM kernel polynomial
3. Decission Tree

Anda boleh melakukan eksplorasi dengan melakukan tunning hyperparameter



**1. Import library tambahan**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

**2. Persiapan Data**
<br>Membaca dataset diabetes.csv dan mempersiapkan data untuk digunakan pada model.

```python
# Memuat dataset diabetes
data = pd.read_csv('data/diabetes.csv')

# Menampilkan 5 data teratas
df.head()
```

```python
# Cek nama kolom
data.columns
```

```python
# Cek kolom null
data.isnull().sum()
```

```python
# Pada kasus ini, agak tidak masuk akal jika beberapa parameter bernilai 0
# sebagai contoh adalah nilai 'Glucose', 'BloodPlessure' ataupun 'Insulin'.
# Sekecil apapun nilainya, setiap manusia yang hidup pasti miliki nilai-nilai tersebut

# Kita akan manipulasi nilai yang 0 dengan melakukan 'imputasi' atau mengganti nilainya dengan nilai sintetis
# Pada kasus ini, kita akan menggunakan nilai mean

# Cek kolom neng nilai 0
feature_columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
for column in feature_columns:
    print("============================================")
    print(f"{column} ==> Missing zeros : {len(data.loc[data[column] == 0])}")
```

```python
# Impute nilai 0 dengan mean
from sklearn.impute import SimpleImputer

fill_values = SimpleImputer(missing_values=0, strategy="mean", copy=False)

data[feature_columns] = fill_values.fit_transform(data[feature_columns])
```

**3. Pembagian Data**
<br>Memisahkan data menjadi data latih (X_train, y_train) dan data uji (X_test, y_test) dengan proporsi 80% latih dan 20% uji.

```python
X = data[feature_columns]
y = data.Outcome

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**4. Normalisasi Data**
<br>Karena SVM sensitif terhadap skala fitur, maka akan dilakukan standarisasi data dengan StandardScaler.

```python
# Normalisasi fitur menggunakan StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**5. Membuat Model Voting Classifier**
<br>Menggunakan tiga model dasar: Logistic Regression, SVM dengan kernel polynomial, dan Decision Tree.

```python
# Definisikan model individu
log_reg = LogisticRegression(random_state=42)
svm = SVC(kernel='poly', probability=True, random_state=42)  # Menggunakan kernel polynomial
dt = DecisionTreeClassifier(random_state=42)

# Membuat VotingClassifier dengan model-model di atas
voting_clf = VotingClassifier(estimators=[('lr', log_reg), ('svc', svm), ('dt', dt)], voting='hard')
```

**6. Hyperparameter Tuning**
<br>Untuk masing-masing model, dilakukan hyperparameter tuning menggunakan GridSearchCV. Pada contoh ini, fokus pada beberapa hyperparameter utama.

```python
# Logistic Regression:

param_grid_lr = {
    'C': [0.1, 1, 10]
}
grid_lr = GridSearchCV(log_reg, param_grid_lr, cv=5, scoring='accuracy')
grid_lr.fit(X_train, y_train)
best_lr = grid_lr.best_estimator_

# Hasil hyperparameter terbaik
print("Best parameters for Logistic Regression:", best_lr)

# Memprediksi dengan model terbaik
y_pred_lr = grid_lr.predict(X_test)

# Menghitung akurasi
acc_lr = accuracy_score(y_test, y_pred_lr)
print("Akurasi Logistic Regression: {:.4f}".format(acc_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
# SVM dengan kernel polynomial:

param_grid_svm = {
    'C': [0.1, 1, 10],
    'degree': [2, 3, 4]  # Derajat polynomial
}
grid_svm = GridSearchCV(svm, param_grid_svm, cv=5, scoring='accuracy')
grid_svm.fit(X_train, y_train)
best_svm = grid_svm.best_estimator_

# Hasil hyperparameter terbaik
print("Best parameters for SVM:", best_svm)

# Memprediksi dengan model terbaik
y_pred_svm = grid_svm.predict(X_test)

# Menghitung akurasi
acc_svm = accuracy_score(y_test, y_pred_svm)
print("Akurasi SVM: {:.4f}".format(acc_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
# Decision Tree:

param_grid_dt = {
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}
grid_dt = GridSearchCV(dt, param_grid_dt, cv=5, scoring='accuracy')
grid_dt.fit(X_train, y_train)
best_dt = grid_dt.best_estimator_

# Hasil hyperparameter terbaik
print("Best parameters for Decision Tree:", best_dt)

# Memprediksi dengan model terbaik
y_pred_dt = grid_dt.predict(X_test)

# Menghitung akurasi
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Akurasi Decision Tree: {:.4f}".format(acc_dt))
print(classification_report(y_test, y_pred_dt))
```

**7. Membangun Model Ensemble**
<br>Menggunakan model terbaik dari masing-masing algoritma setelah GridSearchCV, pada bagian ini membangun model ensemble dengan voting classifier.

```python
# Voting classifier dengan model terbaik dari hasil GridSearchCV
voting_clf_best = VotingClassifier(estimators=[('lr', best_lr), ('svc', best_svm), ('dt', best_dt)], voting='hard')
voting_clf_best.fit(X_train, y_train)
```

**8. Evaluasi Model**
<br>Setelah model dilatih, selanjutnya melakukan prediksi pada data uji dan menghitung akurasi.

```python
# Prediksi pada data uji
y_pred = voting_clf_best.predict(X_test)

# Menghitung akurasi
acc_voting = accuracy_score(y_test, y_pred)
print(f'Akurasi Voting Classifier: {acc_voting:.4f}')

# Menampilkan classification report
print(classification_report(y_test, y_pred))
```
